```python
df = pd.DataFrame(columns=['ticker', 'avg_diff_metrica', 'q5_diff_metrica','q25_diff_metrica'])

today_date = dt.date.today().strftime('%d/%m/%Y')
delta_days = dt.timedelta(40) # сколько дней надо извлечь
second_date = (dt.date.today() - delta_days).strftime('%d/%m/%Y')

ticker_list = trade_data1.ticker.tolist()

for i in range(len(ticker_list)):
    try:
        search_result = investpy.search_quotes(text = ticker_list[i], products=['stocks'], n_results=1)
    except:
        df.loc[i] = [ticker_list[i],'not_found','not_found','not_found']
        continue
    historical_data = search_result.retrieve_historical_data(from_date=second_date, to_date=today_date)
    historical_data = historical_data.drop(columns = ['Volume','Change Pct'])
    historical_data['raise_drop'] = historical_data.Close - historical_data.Open
    historical_data['high_open'] = (historical_data.High / historical_data.Open * 100) - 100
    historical_data['low_open'] = (historical_data.Low / historical_data.Open * 100) - 100
    historical_data['close_open'] = (historical_data.Close / historical_data.Open * 100) - 100
    historical_data_raise = historical_data.query("raise_drop > 0")
    historical_data_raise['diff_metrica'] = historical_data_raise.close_open / historical_data_raise.high_open
    historical_data_raise['diff_metrica_1'] = (historical_data_raise.Close-historical_data_raise.Open) / (historical_data_raise.High-historical_data_raise.Open)
    historical_data_drop = historical_data.query("raise_drop < 0")
    historical_data_drop['diff_metrica'] = historical_data_drop.close_open / historical_data_drop.low_open
    historical_data['diff_metrica'] = historical_data_raise.diff_metrica
    historical_data['diff_metrica'] = historical_data.diff_metrica.fillna(historical_data_drop.diff_metrica)
    avg_diff_metrica = historical_data['diff_metrica'].mean()
    q5_diff_metrica = historical_data['diff_metrica'].quantile(q=0.5)
    q25_diff_metrica = historical_data['diff_metrica'].quantile(q=0.25)
    df.loc[i] = [ticker_list[i],avg_diff_metrica,q5_diff_metrica,q25_diff_metrica]
```

```python
historical_data = historical_data.drop(columns = ['Volume','Change Pct'])
historical_data['raise_drop'] = historical_data.Close - historical_data.Open
historical_data
```

```python
historical_data['high_low'] = historical_data.High - historical_data.Low
historical_data['close_open'] = historical_data.Close - historical_data.Open
historical_data['open_close'] = historical_data.Open - historical_data.Close
```

```python
historical_data_raise = historical_data.query("raise_drop > 0")
historical_data_raise['diff_metrica111'] = historical_data_raise.close_open / historical_data_raise.high_low
```

```python
historical_data_drop = historical_data.query("raise_drop < 0")
historical_data_drop['diff_metrica111'] = historical_data_drop.open_close / historical_data_drop.high_low
```

```python
historical_data['diff_metrica111'] = historical_data_raise.diff_metrica111
historical_data['diff_metrica111'] = historical_data.diff_metrica111.fillna(historical_data_drop.diff_metrica111)
```

```python
avg_diff_metrica = historical_data['diff_metrica111'].mean()
q5_diff_metrica = historical_data['diff_metrica111'].quantile(q=0.5)
q25_diff_metrica = historical_data['diff_metrica111'].quantile(q=0.25)
q75_diff_metrica = historical_data['diff_metrica111'].quantile(q=0.75)
```
